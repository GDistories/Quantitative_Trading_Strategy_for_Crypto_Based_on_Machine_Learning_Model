```python
import requests

# Define the endpoint URL
url = "https://api.binance.us/api/v3/klines"
# 2023/11/01 00:00:00
time_start = 1698768000000
# 2023/11/07 23:59:59
time_end = 1699372799999

# Define the parameters for the API call
params = {
    'symbol': 'BTCUSDT',  # Replace with BTCUSDT for Bitcoin data
    'interval': '4h',  # Replace with the desired interval
    'startTime': time_start,  # Replace with the desired start time in milliseconds
    'endTime': time_end,  # Replace with the desired end time in milliseconds
    'limit': 100000000000  # The maximum number of Klines to retrieve
}

# Make the GET request
response = requests.get(url, params=params)

# Print the response from the API
print(response.json())

#convert json to df
import pandas as pd
df = pd.DataFrame(response.json())
df
```
