```python
df1 = pd.read_csv("Merged.csv",error_bad_lines=False,encoding='utf-8')
df1.drop('text', axis=1, inplace=True)
```

```python
df1['hard_cleaned_text'][0:15]
```

```python
# Get subjectivity:
def get_subjectivity(text):
    blob = TextBlob(str(text))
    return blob.sentiment.subjectivity

# Get polarity:
def get_polarity(text):
    blob = TextBlob(str(text))
    return blob.sentiment.polarity

df1['subjectivity'] = df1['hard_cleaned_text'].apply(get_subjectivity)
df1['polarity'] = df1['hard_cleaned_text'].apply(get_polarity)

df1.head()
#底下有一些colab自动建议生成的图像 但感觉没什么用
```

```python
df2 = pd.read_csv("merged_df.csv")
df2.head()
```

```python
import matplotlib.pyplot as plt

df2 = df2.drop(df2[df2['label'] == 0].index)
# 获取情感倾向值和收盘价的数据
sentiment = df2['label']
closing_price = df2['close']

# 绘制散点图
plt.scatter(sentiment, closing_price)
plt.xlabel('Sentiment')
plt.ylabel('Closing Price')
plt.title('Sentiment vs Closing Price')

# 显示图形
plt.show()

##感觉没啥用
```

```python
#         按月平均的close和label
# 将日期列转换为日期类型
df2['date'] = pd.to_datetime(df2['date'])

# 按月份进行分组，并计算每个月的平均收盘价和平均label值
monthly_data = df2.groupby(df2['date'].dt.to_period('M')).mean()

# 提取月份、平均收盘价和平均label值数据
months = monthly_data.index.to_timestamp()
closing_prices = monthly_data['close']
average_label = monthly_data['label']

# 创建图形和轴对象
fig, ax1 = plt.subplots()

# 绘制柱状图
ax1.plot(months, closing_prices, color='b',marker='o')
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Closing Price')
ax1.set_title('Monthly Average Closing Price')

# 创建第二个y轴，并绘制label数据
ax2 = ax1.twinx()
ax2.plot(months, average_label, color='r', marker='x')
ax2.set_ylabel('Average Label')

# 旋转x轴标签，使其更易读
plt.xticks(rotation=45)

# 显示图形
plt.show()
```

```python
#            按周平均的close和label
# 将日期列转换为日期类型
df2['date'] = pd.to_datetime(df2['date'])

# 将数据按周进行重新采样，并计算每周的平均收盘价和平均label值
weekly_data = df2.resample('W', on='date').mean()

# 提取日期、平均收盘价和平均label值数据
dates = weekly_data.index
closing_prices = weekly_data['close']
average_label = weekly_data['label']

# 设置图形的宽度和高度
fig = plt.figure(figsize=(12, 6))

# 创建图形和轴对象
ax1 = fig.add_subplot(111)

# 绘制柱状图
ax1.plot(dates, closing_prices, color='b', marker='o')
ax1.set_xlabel('Week')
ax1.set_ylabel('Average Closing Price')
ax1.set_title('Weekly Average Closing Price')

# 创建第二个y轴，并绘制label数据
ax2 = ax1.twinx()
ax2.plot(dates, average_label, color='r', marker='x')
ax2.set_ylabel('Average Label')

# 旋转x轴标签，使其更易读
plt.xticks(rotation=45)

# 显示图形
plt.show()
```
